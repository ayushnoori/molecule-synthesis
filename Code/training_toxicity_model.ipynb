```python
from typing import Any, cast

from DeepPurpose import utils, CompoundPred
from tdc.single_pred import Tox
from tdc.utils import retrieve_dataset_names, retrieve_label_name_list

from utils.printing import IPythonPrinter
from multi_model import MultiModel, MultiModelSinglePredictionModelArgument
```

```python
printer = IPythonPrinter()
```

```python
tox_datasets = cast(list[str], retrieve_dataset_names("Tox"))
print(f"Number of Tox datasets available from TDC: {len(tox_datasets)}")
printer.print_markdown("**Datasets available:**")
for dataset in tox_datasets:
    printer.print_markdown(f"1. {dataset}")
printer.flush()
```

```python
tox_datasets_labels: list[tuple[str, str]] = []
for dataset in tox_datasets:
    try:
        labels = retrieve_label_name_list(dataset)
        for label in labels:
            tox_datasets_labels.append((dataset, label))
    except:
        tox_datasets_labels.append((dataset, None))

tox_datasets_labels
```

```python
augmented_models = [
    MultiModelSinglePredictionModelArgument(
        name=f"{dataset_name}_{label_name}",
        model=None,
        dataloader=Tox(name=dataset_name, label_name=label_name, path="../../Data/TDC_Tox"),
    )
    for (dataset_name, label_name) in tox_datasets_labels
]
augmented_models
```

```python
multi_model = MultiModel(augmented_models=augmented_models)
multi_model
```

```python
multi_model.train()
```

```python
multi_model.save(model_dir="../../Data/property_predictors/multi_models/toxicity")
```
